```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('../data/03-20-23.xlsx')
df['epsilon'] = df['epsilon']*1000
```

```python
# delta_list = df['delta'].unique()
# alpha_list = df['alpha'].unique()
# interval_list = df['interval'].unique()
delta_list = [2000]
alpha_list = [5]
interval_list = df['interval'].unique()
epsilon_list = df['epsilon'].unique()
# Effect of epsilon
for delta in delta_list:
    for interval in interval_list:
        for alpha in alpha_list:
                sample_df = df[(df['delta'] == delta) & (df['interval'] == interval) & (df['alpha'] == alpha)]
                if(len(sample_df.index) != 0):
                    plt.plot(sample_df['epsilon'], sample_df['mean_offset_size'])
                    plt.title('Mean offset size: D{delta}.I{interval}.A{alpha}'.format(delta = delta, interval = interval, alpha = alpha))
                    plt.show()
                    # print(sample_df.head())
            # sample_df = df[(df['delta'] == delta) & (df['interval'] == interval) & (df['alpha'] == alpha)]
            # plt.plot(sample_df['epsilon'], sample_df['mean_counter_size'])
            # plt.xlabel('Mean counter size: D{delta}.I{interval}.A{alpha}'.format(delta = delta, interval = interval, alpha = alpha))
            # plt.show()
```
