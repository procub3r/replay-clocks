```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../data/2023-04-20-hvc.csv')
df.head()
```

```python
for epsilon in df['EPSILON'].unique():
    sample_df = df[(df['DELTA'] == 2) & (df['ALPHA'] == 0.025) & (df['EPSILON'] == epsilon)].reset_index(drop=True)
    plt.plot(sample_df.index, sample_df['OFFSIZE'], label = 'E{}'.format(epsilon))
    plt.legend()
    plt.show()
```

```python
sample_df = df[(df['DELTA'] == 8) & (df['ALPHA'] == 0.025)].reset_index(drop=True)
print(sample_df)
plt.plot(sample_df['EPSILON'], sample_df['OFFSIZE'])
plt.legend()
plt.show()
```
