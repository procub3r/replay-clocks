```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../debug/out-2023-05-21.csv')
df.head()
```

```python
sample_df = df[(df['delta'] == 16) & (df['interval'] == 32)]
sample_df
```

```python
fig = plt.figure(figsize=(20, 10))
alpha1 = sample_df[sample_df['alpha'] == 0.25]
alpha2 = sample_df[sample_df['alpha'] == 0.5]
alpha3 = sample_df[sample_df['alpha'] == 1]
# print(alpha3)
plt.plot(alpha1['epsilon'], alpha1['offsize'], label='A.0.25')
plt.plot(alpha2['epsilon'], alpha2['offsize'], label='A.0.50')
plt.plot(alpha3['epsilon'], alpha3['offsize'], label='A.1')
plt.legend()
plt.show()
```
